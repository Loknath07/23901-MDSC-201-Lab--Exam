# **1st Problem**

```r
data <- read_excel("/content/Months_Songs.xlsx")
```

```r
data
```

```r
correlation_coefficient <- cor(data$Months, data$Songs)
```

```r
correlation_coefficient
```

```r
print(colnames(data))
```

```r
# Performing Simple Linear Regression
model <- lm(Songs ~ Months, data)

# Plotting the scatter plot of the data
plot(data$Months, data$Songs, main="Scatter Plot with Fitted Line", xlab="Months", ylab="Songs")

# Add the regression line to the scatter plot
abline(model, col="red")
legend("topright", legend=c("Data", "Fitted Line"), col=c("black", "red"), lty=1:1, cex=0.8)
```
